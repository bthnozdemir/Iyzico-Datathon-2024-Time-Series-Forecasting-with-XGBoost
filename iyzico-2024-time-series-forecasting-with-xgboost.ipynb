```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# EÄŸitim verisini dataframe'e aktar.
df = pd.read_csv('/kaggle/input/iyzico-datathon/train.csv')

# month_id sÃ¼tununu 'datetime' formatÄ±na dÃ¶nÃ¼ÅŸtÃ¼r.
df['month_id'] = pd.to_datetime(df['month_id'], format='%Y%m')


first_unique_merchant_count = df['merchant_id'].nunique()
print(f'Unique Merchant Count: {first_unique_merchant_count}')

# Tarih aralÄ±ÄŸÄ± gÃ¶rselleÅŸtirme
plt.figure(figsize=(12,6))
df['month_id'].hist(bins=50)
plt.title("Veri DaÄŸÄ±lÄ±mÄ± - TÃ¼m Zamanlar")
plt.xlabel("Tarih")
plt.ylabel("Veri SayÄ±sÄ±")
plt.show()
```

```python
#Sonraki 3 ay iÃ§in iÅŸlem sayÄ±sÄ± tahmini yaptÄ±ÄŸÄ±mÄ±z iÃ§in veri setimizin son iki ayÄ±nda iÅŸlem gÃ¶rmemiÅŸ 'merchant_id'leri 'churn' ediyoruz.
# 2023 AÄŸustos veya EylÃ¼l ayÄ±nda verisi olmayan tÃ¼m 'merchant_id' lerin bÃ¼tÃ¼n satÄ±rlarÄ±nÄ± siliyoruz.
#Churn ettiÄŸimiz id'ler iÃ§in tahmin yapÄ±lmayacak ve submission oluÅŸtururken tÃ¼m deÄŸerler '0' olmak Ã¼zere sadece tahmini yapÄ±lan 'merhacnt_id'lerin satÄ±rlarÄ±nÄ± dolduracaÄŸÄ±z.
df = df.groupby('merchant_id').filter(lambda x: pd.Timestamp('2023-08-01') in x['month_id'].values or pd.Timestamp('2023-09-01') in x['month_id'].values)

print("\n3. CHURN ANALÄ°ZÄ°")
initial_count = df['merchant_id'].nunique()
print(f'BaÅŸlangÄ±Ã§taki Unique Merchant: {initial_count}')

# Churn Ã¶ncesi son 2 ay aktivitesi
last_two_months = df[df['month_id'].isin(['2023-08-01', '2023-09-01'])]
active_merchants = last_two_months['merchant_id'].unique()

plt.figure(figsize=(8,4))
sns.countplot(x=last_two_months['month_id'].dt.month)
plt.title("AÄŸustos-EylÃ¼l 2023 Aktivite SayÄ±larÄ±")
plt.xlabel("Ay")
plt.ylabel("Aktif Merchant SayÄ±sÄ±")
plt.show()
```

```python
original_columns = df.columns.tolist()

# DiÄŸer Ã¶zellikler ile birleÅŸik analiz yapabilecek bir bilgim yok bu yÃ¼zden 'net_paymet_count' harici tÃ¼m Ã¶zellikleri siliyoruz.
df = df.drop(columns=['settlement_period', 'working_type', 'mcc_id', 'merchant_segment', 'merchant_source_name'])
```

```python
# merchant_id gruplarÄ±nÄ± tarihe gÃ¶re kÃ¼Ã§Ã¼kten bÃ¼yÃ¼ÄŸe sÄ±rala.
df = df.groupby('merchant_id', group_keys=False).apply(lambda x: x.sort_values('month_id', ascending=True))

# SÄ±ralama kontrolÃ¼ iÃ§in Ã¶rnek bir merchant
sample_merchant = df['merchant_id'].iloc[0]
df[df['merchant_id'] == sample_merchant].plot(
    x='month_id', 
    y='net_payment_count', 
    kind='line',
    title=f"Ã–rnek Merchant ({sample_merchant}) Zaman Serisi"
)
plt.show()
```

```python
# Eksik deÄŸer doldurma: 'net_payment_count' sÃ¼tununda boÅŸ olan tÃ¼m hÃ¼crelere 0 deÄŸeri ver.
df['net_payment_count'] = df['net_payment_count'].fillna(0)


print("\n7. TARÄ°H FÄ°LTRESÄ°")
print(f"Filtre Ã¶ncesi en eski tarih: {df['month_id'].min()}")
df = df[df['month_id'] > pd.Timestamp('2022-09-01')]

plt.figure(figsize=(10,5))
df['month_id'].hist(bins=30)
plt.title("FiltrelenmiÅŸ Veri DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("Tarih")
plt.ylabel("KayÄ±t SayÄ±sÄ±")
plt.show()
```

```python
import time
import pandas as pd
from xgboost import XGBRegressor
import numpy as np

#Tahminleri kaydedeceÄŸimiz bir veri seti oluÅŸturma ve 'result' deÄŸiÅŸkenine aktarma.
result = pd.DataFrame(columns=['merchant_id', '202310', '202311', '202312'])

#Modelin ilerlemesini Ã§Ä±ktÄ± almak gÃ¶rÃ¼ntÃ¼lemek iÃ§in iÅŸlenen 'merchant_id' sayacÄ±.
counter = 0

# Her bir merchant_id iÃ§in sayaÃ§ artsÄ±n ve 87*5 nin katlarÄ±nda yÃ¼zde 1 artÄ±ÅŸ olarak Ã§Ä±ktÄ± versin. 
for merchant_id in df['merchant_id'].unique():
    counter += 1
    if counter % 435 == 0:
      print(f"Ä°lerleme: %{100 * counter / final_count}")

    # Her merchant_id'ye ait toplam veri sayÄ±sÄ±na gÃ¶re dinamik olarak 'window_size' ve 'xgb parametrelerini farklÄ± ayarlamak iÃ§in 'satir_sayisi' deÄŸiÅŸkeni oluÅŸtur
    satir_sayisi = len(df[df['merchant_id'] == merchant_id])
    #'satir_sayisi' yani veri sayÄ±sÄ± bÃ¼yÃ¼klÃ¼ÄŸÃ¼ne gÃ¶re farklÄ± parametreler ile eÄŸitim gerÃ§ekleÅŸtir
    if satir_sayisi > 9:
        window_size = 10
        model = XGBRegressor(objective='reg:squarederror', n_estimators=200, learning_rate=0.010, max_depth=6,gamma=0.3, reg_lambda=1.5)
    elif satir_sayisi > 6:
        window_size = 7
        model = XGBRegressor(objective='reg:squarederror', n_estimators=150, learning_rate=0.015, max_depth=3, gamma=0.2, reg_lambda=1.2)
    elif satir_sayisi > 3:
        window_size = 4
        model = XGBRegressor(objective='reg:squarederror', n_estimators=120, learning_rate=0.02, max_depth=1, gamma=0.1, reg_lambda=1.0)
    else:
        # 3 veya daha az satÄ±r iÃ§in var olan son deÄŸeri sonuÃ§ gir
        next_value = df[df['merchant_id'] == merchant_id]['net_payment_count'].mean()
        new_row = pd.DataFrame({'merchant_id': [f"{merchant_id}"],
                                '202310': [next_value],
                                '202311': [next_value],
                                '202312': [next_value]})
        result = pd.concat([result, new_row], ignore_index=True)
        continue  # DiÄŸer kodlarÄ± atla ve bir sonraki merchant_id'ye geÃ§

    for i in range(window_size, satir_sayisi+1):
        X_train_xgboost = np.array([[j] for j in range(i-window_size+1, i+1)])
        y_train_xgboost = df[df['merchant_id'] == merchant_id]['net_payment_count'].tolist()[i-window_size:i]

        # Modeli eÄŸit
        model.fit(X_train_xgboost, y_train_xgboost)

    # Sonraki Ã¼Ã§ deÄŸeri tahmin ettiÄŸimiz deÄŸerin Ã¼zerine tahmin ederek ilerletme
    next_values = []
    for i in range(satir_sayisi + 1, satir_sayisi + 4):
      # Modeli yeniden eÄŸit
      model.fit(X_train_xgboost, y_train_xgboost)
      next_value = model.predict(np.array([[i]]))[0]
      next_values.append(next_value)
      # EÄŸitim setine tahmin edilen deÄŸeri ekle
      y_train_xgboost.append(next_value)
      X_train_xgboost = np.append(X_train_xgboost, [[i+1]], axis=0)


    # Sonucu kaydet
    new_row = pd.DataFrame({'merchant_id': [f"{merchant_id}"],
                            '202310': [next_values[0]],
                            '202311': [next_values[1]],
                            '202312': [next_values[2]]})
    result = pd.concat([result, new_row], ignore_index=True)

#Veri setinin ilk satÄ±rlarÄ±nÄ± yazdÄ±r
result.head()
```

```python
import matplotlib.pyplot as plt

# Durumlar iÃ§in sayaÃ§lar ve toplamlar
durumlar = {
    "BÃ¼yÃ¼k Veri Seti (>9 satÄ±r)": {"count": 0, "toplam_net_payment": 0},
    "Orta BÃ¼yÃ¼klÃ¼kte Veri Seti (7-9 satÄ±r)": {"count": 0, "toplam_net_payment": 0},
    "KÃ¼Ã§Ã¼k Veri Seti (4-6 satÄ±r)": {"count": 0, "toplam_net_payment": 0},
    "Ã‡ok KÃ¼Ã§Ã¼k Veri Seti (<=3 satÄ±r)": {"count": 0, "toplam_net_payment": 0}
}

# Her bir merchant_id iÃ§in durumlarÄ± hesapla
for merchant_id in df['merchant_id'].unique():
    satir_sayisi = len(df[df['merchant_id'] == merchant_id])
    net_payment_toplam = df[df['merchant_id'] == merchant_id]['net_payment_count'].sum()
    
    if satir_sayisi > 9:
        durumlar["BÃ¼yÃ¼k Veri Seti (>9 satÄ±r)"]["count"] += 1
        durumlar["BÃ¼yÃ¼k Veri Seti (>9 satÄ±r)"]["toplam_net_payment"] += net_payment_toplam
    elif satir_sayisi > 6:
        durumlar["Orta BÃ¼yÃ¼klÃ¼kte Veri Seti (7-9 satÄ±r)"]["count"] += 1
        durumlar["Orta BÃ¼yÃ¼klÃ¼kte Veri Seti (7-9 satÄ±r)"]["toplam_net_payment"] += net_payment_toplam
    elif satir_sayisi > 3:
        durumlar["KÃ¼Ã§Ã¼k Veri Seti (4-6 satÄ±r)"]["count"] += 1
        durumlar["KÃ¼Ã§Ã¼k Veri Seti (4-6 satÄ±r)"]["toplam_net_payment"] += net_payment_toplam
    else:
        durumlar["Ã‡ok KÃ¼Ã§Ã¼k Veri Seti (<=3 satÄ±r)"]["count"] += 1
        durumlar["Ã‡ok KÃ¼Ã§Ã¼k Veri Seti (<=3 satÄ±r)"]["toplam_net_payment"] += net_payment_toplam

# Ortalama net_payment_count hesapla
for durum in durumlar:
    if durumlar[durum]["count"] > 0:
        durumlar[durum]["ortalama_net_payment"] = durumlar[durum]["toplam_net_payment"] / durumlar[durum]["count"]
    else:
        durumlar[durum]["ortalama_net_payment"] = 0

# GÃ¶rselleÅŸtirme
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# 1. Grafik: Durumlara gÃ¶re merchant_id sayÄ±sÄ±
ax[0].bar(durumlar.keys(), [durumlar[durum]["count"] for durum in durumlar], color='skyblue')
ax[0].set_title("Durumlara GÃ¶re Merchant ID SayÄ±sÄ±")
ax[0].set_xlabel("Durum")
ax[0].set_ylabel("Merchant ID SayÄ±sÄ±")
ax[0].tick_params(axis='x', rotation=45)

# 2. Grafik: Durumlara gÃ¶re ortalama net_payment_count
ax[1].bar(durumlar.keys(), [durumlar[durum]["ortalama_net_payment"] for durum in durumlar], color='lightgreen')
ax[1].set_title("Durumlara GÃ¶re Ortalama Net Payment Count")
ax[1].set_xlabel("Durum")
ax[1].set_ylabel("Ortalama Net Payment Count")
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# DurumlarÄ± yazdÄ±r
for durum in durumlar:
    print(f"{durum}:")
    print(f"  Merchant ID SayÄ±sÄ±: {durumlar[durum]['count']}")
    print(f"  Ortalama Net Payment Count: {durumlar[durum]['ortalama_net_payment']:.2f}")
    print()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Her bir merchant_id iÃ§in toplam net_payment_count hesapla
toplam_net_payment = df.groupby('merchant_id')['net_payment_count'].sum()

# Toplam net_payment_count'a gÃ¶re en bÃ¼yÃ¼k 5 merchant_id'yi seÃ§
top_merchants = toplam_net_payment.nlargest(5).index.tolist()

# Her bir 'mechant_id' iÃ§in gÃ¶rselleÅŸtirme
for merchant_id in top_merchants:
    # ID'ye ait verileri filtrele
    merchant_data = df[df['merchant_id'] == merchant_id]
    
    # Result veri setinden tahminleri bul
    next_values = result[result['merchant_id'] == str(merchant_id)].iloc[:, 1:].values.flatten().tolist()
    
    # Regresyon Ã§izgisi iÃ§in hazÄ±rlÄ±k
    X = np.arange(len(merchant_data)).reshape(-1, 1)  # Zaman indeksi
    y = merchant_data['net_payment_count'].values    # GerÃ§ek deÄŸerler
    
    # Lineer regresyon modeli
    regressor = LinearRegression()
    regressor.fit(X, y)
    regression_line = regressor.predict(X)
    
    # Ã‡izim alanÄ± oluÅŸturma
    plt.figure(figsize=(10, 6))
    
    # GerÃ§ek deÄŸerler
    plt.plot(merchant_data['month_id'], 
             merchant_data['net_payment_count'], 
             marker='o', label='GerÃ§ek DeÄŸerler')
    
    # Regresyon Ã§izgisi
    plt.plot(merchant_data['month_id'], 
             regression_line, 
             color='green', linestyle='--', label='Regresyon Ã‡izgisi')
    
    # Tahmin edilen tarihler
    future_dates = pd.date_range(start=merchant_data['month_id'].max(), periods=3, freq='M')
    
    # Tahminleri Ã§iz
    plt.plot(future_dates, next_values, marker='o', color='red', label='Tahminler')
    
    # Grafik ayarlarÄ±
    plt.title(f"{merchant_id} iÃ§in - GerÃ§ek, Regresyon ve Tahmin DeÄŸerleri")
    plt.xlabel("Tarih")
    plt.ylabel("Net Payment Count")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
import time
import pandas as pd
from xgboost import XGBRegressor
import numpy as np

#Tahminleri kaydedeceƒüimiz bir veri seti olu≈üturma ve 'result' deƒüi≈ükenine aktarma.
result = pd.DataFrame(columns=['merchant_id', '202310', '202311', '202312'])

#Modelin ilerlemesini √ßƒ±ktƒ± almak g√∂r√ºnt√ºlemek i√ßin i≈ülenen 'merchant_id' sayacƒ±.
counter = 0

# Her bir merchant_id i√ßin saya√ß artsƒ±n ve 870 in katlarƒ±nda y√ºzde 10 artƒ±≈ü olarak √ßƒ±ktƒ± versin.
for merchant_id in df['merchant_id'].unique():
    counter += 1
    if counter % 870 == 0:
      print(f"ƒ∞lerleme: %{100 * counter / final_count}")

    # Her merchant_id'ye ait toplam veri sayƒ±sƒ±na g√∂re dinamik olarak 'window_size' ve 'xgb parametrelerini farklƒ± ayarlamak i√ßin 'satir_sayisi' deƒüi≈ükeni olu≈ütur
    satir_sayisi = len(df[df['merchant_id'] == merchant_id])
    #'satir_sayisi' yani veri sayƒ±sƒ± b√ºy√ºkl√ºƒü√ºne g√∂re farklƒ± parametreler ile eƒüitim ger√ßekle≈ütir
    if satir_sayisi > 9:
        window_size = 10
        model = XGBRegressor(objective='reg:squarederror', n_estimators=200, learning_rate=0.010, max_depth=6,gamma=0.3, reg_lambda=1.5)
    elif satir_sayisi > 6:
        window_size = 7
        model = XGBRegressor(objective='reg:squarederror', n_estimators=150, learning_rate=0.015, max_depth=3, gamma=0.2, reg_lambda=1.2)
    elif satir_sayisi > 3:
        window_size = 4
        model = XGBRegressor(objective='reg:squarederror', n_estimators=120, learning_rate=0.02, max_depth=1, gamma=0.1, reg_lambda=1.0)
    else:
        # 3 veya daha az satƒ±r i√ßin var olan son deƒüeri sonu√ß gir
        next_value = df[df['merchant_id'] == merchant_id]['net_payment_count'].mean()
        new_row = pd.DataFrame({'merchant_id': [f"{merchant_id}"],
                                '202310': [next_value],
                                '202311': [next_value],
                                '202312': [next_value]})
        result = pd.concat([result, new_row], ignore_index=True)
        continue  # Diƒüer kodlarƒ± atla ve bir sonraki merchant_id'ye ge√ß

    for i in range(window_size, satir_sayisi+1):
        X_train_xgboost = np.array([[j] for j in range(i-window_size+1, i+1)])
        y_train_xgboost = df[df['merchant_id'] == merchant_id]['net_payment_count'].tolist()[i-window_size:i]

        # Modeli eƒüit
        model.fit(X_train_xgboost, y_train_xgboost)

    # Sonraki √º√ß deƒüeri tahmin ettiƒüimiz deƒüerin √ºzerine tahmin ederek ilerletme
    next_values = []
    for i in range(satir_sayisi + 1, satir_sayisi + 4):
      # Modeli yeniden eƒüit
      model.fit(X_train_xgboost, y_train_xgboost)
      next_value = model.predict(np.array([[i]]))[0]
      next_values.append(next_value)
      # Eƒüitim setine tahmin edilen deƒüeri ekle
      y_train_xgboost.append(next_value)
      X_train_xgboost = np.append(X_train_xgboost, [[i+1]], axis=0)


    # Sonucu kaydet
    new_row = pd.DataFrame({'merchant_id': [f"{merchant_id}"],
                            '202310': [next_values[0]],
                            '202311': [next_values[1]],
                            '202312': [next_values[2]]})
    result = pd.concat([result, new_row], ignore_index=True)

#Veri setinin ilk satƒ±rlarƒ±nƒ± yazdƒ±r
result.head()
```

```python
import matplotlib.pyplot as plt

# Durumlar i√ßin saya√ßlar ve toplamlar
durumlar = {
    "B√ºy√ºk Veri Seti (>9 satƒ±r)": {"count": 0, "toplam_net_payment": 0},
    "Orta B√ºy√ºkl√ºkte Veri Seti (7-9 satƒ±r)": {"count": 0, "toplam_net_payment": 0},
    "K√º√ß√ºk Veri Seti (4-6 satƒ±r)": {"count": 0, "toplam_net_payment": 0},
    "√áok K√º√ß√ºk Veri Seti (<=3 satƒ±r)": {"count": 0, "toplam_net_payment": 0}
}

# Her bir merchant_id i√ßin durumlarƒ± hesapla
for merchant_id in df['merchant_id'].unique():
    satir_sayisi = len(df[df['merchant_id'] == merchant_id])
    net_payment_toplam = df[df['merchant_id'] == merchant_id]['net_payment_count'].sum()

    if satir_sayisi > 9:
        durumlar["B√ºy√ºk Veri Seti (>9 satƒ±r)"]["count"] += 1
        durumlar["B√ºy√ºk Veri Seti (>9 satƒ±r)"]["toplam_net_payment"] += net_payment_toplam
    elif satir_sayisi > 6:
        durumlar["Orta B√ºy√ºkl√ºkte Veri Seti (7-9 satƒ±r)"]["count"] += 1
        durumlar["Orta B√ºy√ºkl√ºkte Veri Seti (7-9 satƒ±r)"]["toplam_net_payment"] += net_payment_toplam
    elif satir_sayisi > 3:
        durumlar["K√º√ß√ºk Veri Seti (4-6 satƒ±r)"]["count"] += 1
        durumlar["K√º√ß√ºk Veri Seti (4-6 satƒ±r)"]["toplam_net_payment"] += net_payment_toplam
    else:
        durumlar["√áok K√º√ß√ºk Veri Seti (<=3 satƒ±r)"]["count"] += 1
        durumlar["√áok K√º√ß√ºk Veri Seti (<=3 satƒ±r)"]["toplam_net_payment"] += net_payment_toplam

# Ortalama net_payment_count hesapla
for durum in durumlar:
    if durumlar[durum]["count"] > 0:
        durumlar[durum]["ortalama_net_payment"] = durumlar[durum]["toplam_net_payment"] / durumlar[durum]["count"]
    else:
        durumlar[durum]["ortalama_net_payment"] = 0

# G√∂rselle≈ütirme
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# 1. Grafik: Durumlara g√∂re merchant_id sayƒ±sƒ±
ax[0].bar(durumlar.keys(), [durumlar[durum]["count"] for durum in durumlar], color='skyblue')
ax[0].set_title("Durumlara G√∂re Merchant ID Sayƒ±sƒ±")
ax[0].set_xlabel("Durum")
ax[0].set_ylabel("Merchant ID Sayƒ±sƒ±")
ax[0].tick_params(axis='x', rotation=45)

# 2. Grafik: Durumlara g√∂re ortalama net_payment_count
ax[1].bar(durumlar.keys(), [durumlar[durum]["ortalama_net_payment"] for durum in durumlar], color='lightgreen')
ax[1].set_title("Durumlara G√∂re Ortalama Net Payment Count")
ax[1].set_xlabel("Durum")
ax[1].set_ylabel("Ortalama Net Payment Count")
ax[1].tick_params(axis='x', rotation=45)

plt.tight_layout()
plt.show()

# Durumlarƒ± yazdƒ±r
for durum in durumlar:
    print(f"{durum}:")
    print(f"  Merchant ID Sayƒ±sƒ±: {durumlar[durum]['count']}")
    print(f"  Ortalama Net Payment Count: {durumlar[durum]['ortalama_net_payment']:.2f}")
    print()
```

```python
from sklearn.linear_model import LinearRegression
import numpy as np

# Her bir merchant_id i√ßin toplam net_payment_count hesapla
toplam_net_payment = df.groupby('merchant_id')['net_payment_count'].sum()

# Toplam net_payment_count'a g√∂re en b√ºy√ºk 5 merchant_id'yi se√ß
top_merchants = toplam_net_payment.nlargest(5).index.tolist()

# Her bir 'mechant_id' i√ßin g√∂rselle≈ütirme
for merchant_id in top_merchants:
    # ID'ye ait verileri filtrele
    merchant_data = df[df['merchant_id'] == merchant_id]

    # Result veri setinden tahminleri bul
    next_values = result[result['merchant_id'] == str(merchant_id)].iloc[:, 1:].values.flatten().tolist()

    # Regresyon √ßizgisi i√ßin hazƒ±rlƒ±k
    X = np.arange(len(merchant_data)).reshape(-1, 1)  # Zaman indeksi
    y = merchant_data['net_payment_count'].values    # Ger√ßek deƒüerler

    # Lineer regresyon modeli
    regressor = LinearRegression()
    regressor.fit(X, y)
    regression_line = regressor.predict(X)

    # √áizim alanƒ± olu≈üturma
    plt.figure(figsize=(10, 6))

    # Ger√ßek deƒüerler
    plt.plot(merchant_data['month_id'],
             merchant_data['net_payment_count'],
             marker='o', label='Ger√ßek Deƒüerler')

    # Regresyon √ßizgisi
    plt.plot(merchant_data['month_id'],
             regression_line,
             color='green', linestyle='--', label='Regresyon √áizgisi')

    # Tahmin edilen tarihler
    future_dates = pd.date_range(start=merchant_data['month_id'].max(), periods=3, freq='M')

    # Tahminleri √ßiz
    plt.plot(future_dates, next_values, marker='o', color='red', label='Tahminler')

    # Grafik ayarlarƒ±
    plt.title(f"{merchant_id} i√ßin - Ger√ßek, Regresyon ve Tahmin Deƒüerleri")
    plt.xlabel("Tarih")
    plt.ylabel("Net Payment Count")
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()
```
